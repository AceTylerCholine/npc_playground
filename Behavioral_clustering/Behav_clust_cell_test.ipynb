```python
def sort_cluster_index_ranges(cluster_index_ranges_dict):
    """Sorts the ranges within a cluster_index_ranges_dict.

    Args:
        cluster_index_ranges_dict: A dictionary where keys are cluster IDs and values
            are lists of (start, end) tuples representing ranges.

    Returns:
        A new dictionary with the same keys but sorted ranges.
    """

    sorted_ranges_dict = {}
    for cluster_id, ranges in cluster_index_ranges_dict.items():
        sorted_ranges_dict[cluster_id] = sorted(ranges)
    return sorted_ranges_dict

# Example Usage (assuming you have the corrected 'cluster_index_ranges' variable)
sorted_ranges_dict = sort_cluster_index_ranges(cluster_index_ranges)

# To get the sorted ranges for cluster ID 7:
sorted_ranges_for_7 = sorted_ranges_dict[7]
print(sorted_ranges_for_7) 
```

```python
def extract_cluster_data(df, cluster_type="cluster_timestamps_ranges_dict"):
    """Extracts cluster data from a DataFrame.

    Args:
        df: The input DataFrame.
        cluster_type: The column name containing the cluster data. 
                      Defaults to 'trial_cluster_timestamps_ranges_dict'.

    Returns:
        A new DataFrame with columns: 'recording', 'subject', 'cluster', and 'ranges'.
    """
    new_df_data = []

    for _, row in df.iterrows():
        base_session_dir = row['session_dir'].split('_subj_')[0]
        subject = row['current_subject'].replace('.', '-')
        recording = f"{base_session_dir}_subj_{subject}"
        
        # Fetch and ensure cluster data is a dictionary
        cluster_data = row[cluster_type]
        if not isinstance(cluster_data, dict):
            print(f"Warning: Invalid cluster data for {recording}: {cluster_data}")
            continue  # Skip this row if data is invalid

        for cluster_id, ranges in cluster_data.items():
            new_df_data.append({
                'recording': recording,
                'subject': subject,
                'cluster': int(cluster_id),
                'ranges': ranges,  # Store the ranges as a list of tuples
            })

    return pd.DataFrame(new_df_data)
    


# Example usage
new_df = extract_cluster_data(df)

# Optionally, sort by recording and cluster to organize the data
new_df = new_df.sort_values(by=['recording', 'cluster'])

new_df
```

```python
import pandas as pd
import numpy as np

def extract_cluster_data(df, cluster_type="cluster_timestamps_ranges_dict"):
    """Extracts cluster data from a DataFrame and reformats it into a structured DataFrame.

    Args:
        df: The input DataFrame.
        cluster_type: The column name containing the cluster data. Defaults to 'cluster_timestamps_ranges_dict'.

    Returns:
        A new DataFrame with columns: 'recording', 'subject', 'cluster', and 'ranges'.
    """
    new_df_data = []

    for _, row in df.iterrows():
        # Construct the base recording string from session_dir and subject
        base_session_dir = row['session_dir'].split('_subj_')[0]
        subject_formatted = row['current_subject'].replace('.', '-')
        recording = f"{base_session_dir}_subj_{subject_formatted}"

        # Fetch and validate cluster data as a dictionary
        cluster_data = row[cluster_type]
        if not isinstance(cluster_data, dict):
            print(f"Warning: Invalid cluster data for {recording}: {cluster_data}")
            continue

        # Append data for each cluster to the new DataFrame list
        for cluster_id, ranges in cluster_data.items():
            new_df_data.append({
                'recording': recording,
                'subject': subject_formatted,
                'cluster': int(cluster_id),
                'ranges': np.array(ranges, dtype=np.int64)  # Convert ranges to numpy arrays for efficient processing
            })

    return pd.DataFrame(new_df_data)

# Usage example (assuming df is already defined)
new_df = extract_cluster_data(df)

new_df
```
