```python
import pandas as pd

# Load the DataFrame using pandas
rce3_alone_clusters = pd.read_pickle("rce_pilot_3_alone_comp_cluster_ranges.pkl")
```

```python
rce3_alone_clusters = pd.read_pickle("rce_pilot_3_alone_comp_cluster_ranges.pkl")
```

```python
updated_rce3_alone_clusters = pd.read_pickle("updated_rce_pilot_3_alone_comp_cluster_ranges.pkl")
```

```python
are_identical = rce3_alone_clusters.equals(updated_rce3_alone_clusters)
print("Dataframes are identical:", are_identical)
```

```python
differences_in_cluster_index = rce3_alone_clusters['cluster_index_ranges_dict'] != updated_rce3_alone_clusters['cluster_index_ranges_dict']
print(rce3_alone_clusters[differences_in_cluster_index][['video_name', 'current_subject', 'cluster_index_ranges_dict']])
```

```python
# Set indices
rce3_indexed = rce3_alone_clusters.set_index(['video_name', 'current_subject'])
updated_rce3_indexed = updated_rce3_alone_clusters.set_index(['video_name', 'current_subject'])

# Compare
differences = rce3_indexed.compare(updated_rce3_indexed)
print(differences)
```

```python
merged_df = pd.merge(rce3_alone_clusters, updated_rce3_alone_clusters, on=['video_name', 'current_subject'], how='outer', indicator=True)
print(merged_df[merged_df['_merge'] != 'both'])
```

```python
comparison_result = rce3_alone_clusters.astype(str).merge(updated_rce3_alone_clusters.astype(str), 
                                                          left_index=True, 
                                                          right_index=True, 
                                                          suffixes=('_old', '_new'), 
                                                          indicator=True)
changes = comparison_result[comparison_result.filter(regex='_old$').columns] != comparison_result[comparison_result.filter(regex='_new$').columns]
print(comparison_result[changes.any(axis=1)])
```
