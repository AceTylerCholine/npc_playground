# Visualization of how z-score histograms should not be normally distributed if there are more than 1 origin

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(0)

# Generate the first array of 100 numbers with mean 8 and standard deviation 1
array1 = np.random.normal(loc=8, scale=1, size=100)

# Generate the second array of 50 numbers with mean 11 and standard deviation 1
array2 = np.random.normal(loc=11, scale=1, size=50)

# Combine both arrays
combined_array = np.concatenate((array1, array2))

# Create a histogram of the combined array
plt.hist(combined_array, bins=20, color='blue', alpha=0.7)
plt.title('Histogram of Combined Normal Distributions')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Generate the first array of 100 numbers with mean 8 and standard deviation 1
array1 = np.random.normal(loc=8, scale=1, size=1000)

# Generate the second array of 50 numbers with mean 11 and standard deviation 1
array2 = np.random.normal(loc=12, scale=1, size=400)

array3 = np.random.normal(loc=5, scale=1, size=300)

# Combine both arrays
combined_array = np.concatenate((array1, array2, array3))

# Create a histogram of the combined array
plt.hist(combined_array, bins=20, color='blue', alpha=0.7)
plt.title('Histogram of Combined Normal Distributions')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Simulated firing rates array (replace this with actual data if available)
firing_rates = np.random.normal(20, 5, 1000)  # mean=20, SD=5, 1000 samples

# Calculate mean and standard deviation
mean_firing_rate = np.mean(firing_rates)
std_firing_rate = np.std(firing_rates)

# Sample with firing rate of 40
sample_rate = 40

# Calculate z-score
z_score = (sample_rate - mean_firing_rate) / std_firing_rate
print(f"Mean Firing Rate: {mean_firing_rate:.2f}")
print(f"Standard Deviation of Firing Rates: {std_firing_rate:.2f}")
print(f"Z-Score of a Firing Rate of 40: {z_score:.2f}")
```
