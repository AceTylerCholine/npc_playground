```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the DataFrame from the pickle file
umap_df = pd.read_pickle('umap_df.pkl')

# Calculate average waveforms for cluster_id 5 and 3
average_waveform_5 = umap_df[umap_df['cluster_id'] == 5]['waveform'].apply(np.array).mean(axis=0)
average_waveform_3 = umap_df[umap_df['cluster_id'] == 3]['waveform'].apply(np.array).mean(axis=0)

# Extract the cluster colors
color_5 = umap_df[umap_df['cluster_id'] == 5]['cluster_color'].iloc[0]
color_3 = umap_df[umap_df['cluster_id'] == 3]['cluster_color'].iloc[0]

# Plot the average waveforms
plt.figure(figsize=(10, 6))
plt.plot(average_waveform_5, label='Cluster 5 Average', color=color_5)
plt.plot(average_waveform_3, label='Cluster 3 Average', color=color_3)

# Find specific rows from cluster 3 where x < 2
specific_rows = umap_df[(umap_df['cluster_id'] == 3) & (umap_df['x'] < 2)]

# Colors for specific rows
specific_colors = [[0.188, 0.949, 0.271, 1.0], [0.51, 0.949, 0.188, 1.0]]

# Plot these specific waveforms with specified colors
for idx, (row, color) in enumerate(zip(specific_rows.iterrows(), specific_colors)):
    idx, row = row
    plt.plot(row['waveform'], label=f'Cluster 3, x < 2 (Row {idx})', color=color)

# Adding labels and legend
plt.title('Average and Specific Waveforms for Cluster IDs 3 and 5')
plt.xlabel('Time')
plt.ylabel('Waveform Amplitude')
plt.legend()

# Show plot
plt.show()
```

```python
umap_df_detail = pd.read_pickle('umap_df_detail.pkl')

# Function to split the recording name
def split_recording_name(name):
    date = name[:8]  # First 8 characters are the date
    subj_index = name.index('_subj_') + 6  # Start of subject
    subj = name[subj_index:subj_index+3].replace('.', '-')  # Next 3 characters after '_subj_' and replace dots

    # Determine the end index for the experiment name (either at '_subj_' or '_D')
    exp_end_index = min(name.find('_subj_'), name.find('_D')) if name.find('_D') != -1 else name.find('_subj_')
    exp = name[16:exp_end_index]  # From the 17th character up to the determined end

    return date, exp, subj

# Apply the function and assign the results to new columns
umap_df_detail[['date', 'experiment', 'subj']] = umap_df_detail['recording_name'].apply(
    lambda x: pd.Series(split_recording_name(x))
)

umap_df_detail
```

```python
# Group by cluster_id and subj, count the occurrences
subj_counts = umap_df_detail.groupby(['cluster_id', 'subj']).size()

# Group by cluster_id to find the total counts per cluster
total_counts = umap_df_detail.groupby('cluster_id').size()

# Calculate the ratio of each subject within each cluster
subj_ratios = subj_counts / total_counts.reindex(subj_counts.index, level=0)

# Convert the Series to a DataFrame for better readability and export if necessary
subj_ratios_df = subj_ratios.reset_index(name='ratio')

subj_ratios_df
```
