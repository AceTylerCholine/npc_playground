```python
import pandas as pd
import numpy as np
import ast
import pickle
from pathlib import Path
from matplotlib_venn import venn2
from matplotlib_venn import venn3
from scipy.stats import gaussian_kde
from scipy.signal import argrelextrema
import matplotlib.pyplot as plt
import seaborn as sns
import spikeanal as spike

# It takes several steps to create the timestamp_dicts, refer to Ephys_Analysis_Notebook to create
timestamp_dicts = pickle.load(open('timestamp_dicts.pkl', 'rb'))

# Construct the path in a platform-independent way (HiPerGator or Windows)
ephys_path = Path('.') / 'export' / 'updated_phys' / 'non-novel' / 'all_non_novel'

ephys_data = spike.EphysRecordingCollection(str(ephys_path))
```

```python
for recording in ephys_data.collection.keys():
    # Check if the recording key (without everything after subject #) is in timestamp_dicts
    start_pos = recording.find('subj_')
    # Add the length of 'subj_' and 3 additional characters to include after 'subj_'
    end_pos = start_pos + len('subj_') + 3
    # Slice the recording key to get everything up to and including the subject identifier plus three characters
    recording_key_without_suffix = recording[:end_pos]
    if recording_key_without_suffix in timestamp_dicts:
        # Assign the corresponding timestamp_dicts dictionary to event_dict
        ephys_data.collection[recording].event_dict = timestamp_dicts[recording_key_without_suffix]
        
        # Extract the subject from the recording key
        start = recording.find('subj_') + 5  # Start index after 'subj_'
        subject = recording[start:start+3]
        
        # Assign the extracted subject
        ephys_data.collection[recording].subject = subject
        
spike_analysis = spike.SpikeAnalysis_MultiRecording(ephys_data, timebin = 100, smoothing_window=250, ignore_freq = 0.5)
```

```python
# Create df of spike freq 10s before and 10s during every event
event_spikes = spike_analysis.generate_event_dataframe()

# Identify 'Pre-event' and 'Event' columns
pre_event_cols = [col for col in event_spikes.columns if col.startswith('Pre-event')]
event_cols = [col for col in event_spikes.columns if col.startswith('Event') and col not in ['Event name', 'Event number']]

# Calculate 'Pre-event M' and 'Pre-event Std' for each row
pre_event_mean = event_spikes[pre_event_cols].mean(axis=1)
pre_event_std = event_spikes[pre_event_cols].std(axis=1).round(3)

# Z-Score 'Event' columns based on 'Pre-event M' and 'Pre-event Std'
normalized_event_cols = event_spikes[event_cols].apply(lambda x: (x - pre_event_mean) / pre_event_std, axis=0)
normalized_event_cols = normalized_event_cols.replace([np.inf, -np.inf], np.nan)  # Handle division by zero by replacing infinities with NaN

# Calculate the average Z-Score for each row
event_z_score = normalized_event_cols.mean(axis=1)

# Create the final DataFrame with the calculated columns at the end
event_spikes_zscores = pd.concat([event_spikes, normalized_event_cols, pre_event_mean.rename('Pre-event M'), pre_event_std.rename('Pre-event Std'), event_z_score.rename('Event Z-Score')], axis=1)

# To prevent issues with further caclulations, you can remove events with zero std.
# This loses ~0.7% of events, but the rest of the event z-scores will be averaged together anyways.
event_spikes_zscores_final = event_spikes_zscores.dropna()

# Average z-scores for each unique recording, unit, and event, ignoring events that only occurred once
filtered_groups = event_spikes_zscores_final.groupby(['Recording', 'Event name', 'Unit number']).filter(lambda x: len(x) >= 2)
zscore_df = filtered_groups.groupby(['Recording', 'Event name', 'Unit number'])['Event Z-Score'].mean().reset_index(name='Z-score')

# Optional: Pickle the zscore_df so you don't have to do these steps again 
pickle.dump(zscore_df, open("zscore_df.pkl", "wb"))

# Print unique events
unique_event_names = zscore_df['Event name'].unique()
print("Unique Event Names:", unique_event_names)
```

```python
# Assuming zscore_df is correctly loaded and contains the 'Z-score' column
kde = gaussian_kde(zscore_df['Z-score'])  # Create KDE from Z-scores
x = np.linspace(-10, 15, 1000)  # Generate x values from -10 to 15 for the KDE evaluation
kde_values = kde(x)  # Evaluate the KDE over the range of x
kde_derivative = np.gradient(kde_values, x)  # Calculate the first derivative of the KDE
kde_second_derivative = np.gradient(kde_derivative, x)  # Calculate the second derivative of the KDE

# Find local minima and maxima of the second derivative
second_derivative_sign_changes = np.diff(np.sign(np.gradient(kde_second_derivative, x)))
local_minima_maxima = x[:-1][np.where(second_derivative_sign_changes != 0)[0]]

# Assuming kde and derivatives have been calculated as per your prior setup
fig, ax = plt.subplots(3, 1, figsize=(10, 12))

# Plot histogram and KDE on different y-axes for the first plot
hist = sns.histplot(zscore_df['Z-score'], ax=ax[0], stat='count', alpha=0.1, edgecolor=(0, 0, 0, 0.3))
secondary_ax = ax[0].twinx()
secondary_ax.plot(x, kde_values, label='Kernel Density Estimation')
secondary_ax.set_ylabel('Density')  # Label for the second y-axis
ax[0].set_title('Kernel Density Estimation with Histogram')
ax[0].vlines(local_minima_maxima, ymin=0, ymax=ax[0].get_ylim()[1], color='grey', linestyle='--', alpha=0.5, label='2nd Derivative Local Min/Max')
secondary_ax.legend()

# Set consistent x-axis limits for the KDE plot
ax[0].set_xlim(-10, 15)

# Plot first derivative of KDE on the second plot
ax[1].plot(x, kde_derivative, label='KDE Derivative')
ax[1].set_title('First Derivative of KDE')
ax[1].hlines(0, x.min(), x.max(), color='red', linestyle='--')
ax[1].vlines(local_minima_maxima, ymin=min(kde_derivative), ymax=max(kde_derivative), color='grey', linestyle='--', alpha=0.5, label='2nd Derivative Local Min/Max')
ax[1].legend()
ax[1].set_ylabel('Density')  # Label for y-axis
ax[1].set_xlim(-10, 15)

# Plot second derivative of KDE on the third plot
ax[2].plot(x, kde_second_derivative, label='KDE 2nd Derivative')
ax[2].set_title('Second Derivative of KDE')
ax[2].hlines(0, x.min(), x.max(), color='red', linestyle='--')
ax[2].vlines(local_minima_maxima, ymin=min(kde_second_derivative), ymax=max(kde_second_derivative), color='grey', linestyle='--', alpha=0.5, label='2nd Derivative Local Min/Max')
ax[2].legend()
ax[2].set_ylabel('Density')  # Label for y-axis
ax[2].set_xlim(-10, 15)

plt.tight_layout()
# plt.show()
plt.savefig('zscore_histogram_derivatives.png')

print("Local minima and maxima from second derivative:", local_minima_maxima)
```

```python
import numpy as np
import pandas as pd
from scipy.stats import gaussian_kde

# Assuming zscore_df is correctly loaded and contains the 'Z-score' column
# Define zscore_df DataFrame here

# Unique event names
unique_event_names = ['lose', 'rewarded', 'win', 'omission', 'both_rewarded']

# Initialize DataFrame to collect results
results_df = pd.DataFrame(columns=['Event', 'Local Min/Max'])

# Define x range for KDE evaluation
x = np.linspace(-10, 15, 1000)

# Process each type of event
for event in unique_event_names:
    event_data = zscore_df[zscore_df['Event name'] == event]['Z-score']
    
    # Create KDE from Z-scores
    kde = gaussian_kde(event_data)
    kde_values = kde(x)  # Evaluate the KDE over the range of x
    
    # Calculate the first and second derivatives of the KDE
    kde_derivative = np.gradient(kde_values, x)
    kde_second_derivative = np.gradient(kde_derivative, x)
    
    # Find local minima and maxima of the second derivative
    second_derivative_sign_changes = np.diff(np.sign(kde_second_derivative))
    local_minima_maxima = x[:-1][np.where(second_derivative_sign_changes != 0)[0]]
    
    # Filter local minima and maxima within specified ranges
    filtered_minmax = [value for value in local_minima_maxima if (2 < value < 7) or (-7 < value < -2)]

    # Append results to DataFrame
    # Create a DataFrame of filtered results for current event and concatenate it
    current_results = pd.DataFrame({
        'Event': [event] * len(filtered_minmax),
        'Local Min/Max': filtered_minmax
    })
    results_df = pd.concat([results_df, current_results], ignore_index=True)

results_df
```
