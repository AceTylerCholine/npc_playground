```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate some time-series data
np.random.seed(0)
data = np.cumsum(np.random.randn(10000))  # Random walk data

def analyze_timebin_size(data, max_bin_size):
    results = {}
    for bin_size in range(1, max_bin_size + 1):
        binned_data = np.mean(data[:len(data) // bin_size * bin_size].reshape(-1, bin_size), axis=1)
        sd = np.std(binned_data)
        mean = np.mean(binned_data)
        results[bin_size] = (mean, sd)
    
    return results

# Analyze different bin sizes
max_bin_size = 200
results = analyze_timebin_size(data, max_bin_size)

# Plotting
bin_sizes = list(results.keys())
sds = [results[size][1] for size in bin_sizes]
plt.figure(figsize=(10, 5))
plt.plot(bin_sizes, sds, marker='o')
plt.xlabel('Timebin Size')
plt.ylabel('Standard Deviation of Binned Data')
plt.title('Effect of Timebin Size on Data Variability')
plt.grid(True)
plt.show()
```
